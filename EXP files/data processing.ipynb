```python
import xml.etree.ElementTree as ET
import csv

# Path to your .shp.xml file
import csv
xml_file = r"C:\Users\Sejal Hanmante\Downloads\Data_for_hackathon\Data_for_hackathon\Stale_Petrol_station\IDN_Q223_May23_Jakarta\Probe.shp.xml"

csv_file = 'output_metadata.csv'

# Parse the XML file
tree = ET.parse(xml_file)
root = tree.getroot()

# Open CSV file for writing
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write header row (modify this based on XML structure)
    writer.writerow(['Tag', 'Attributes', 'Text'])
    
    # Iterate through XML structure and extract relevant data
    for elem in root.iter():
        tag = elem.tag
        attributes = str(elem.attrib)
        text = elem.text.strip() if elem.text else ''
        
        # Write row to CSV
        writer.writerow([tag, attributes, text])

print(f"Data successfully extracted to {csv_file}")
```
